```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from pathlib import Path
```

```python
matplotlib.use("pgf")
matplotlib.rcParams.update(
    {
        "pgf.texsystem": "lualatex",
        "font.family": "serif",
        "font.serif": [],
        "text.usetex": True,
        "pgf.rcfonts": False,
    }
)
```

```python
df= pd.read_csv("/home/andrew/Documents/Sorbonne/M2_1/HPCA/result.csv",index_col= False)
```

```python
df['dt']
```

κ = 0.6, θ = 0.110, σ = 0.3
for these tests, we also fixe ρ = 0.3

Print for fluctuating kappa
 

```python
df_kappa= df[(df["theta"]==0.110) & (df["sigma"]==0.3) & (df["rho"]== 0.3) & (df['dt']<=0.0002)]
df_kappaEuler= df_kappa[df_kappa["euler"] == 1]
df_kappaAlmost= df_kappa[df_kappa["almost_exact"] == 1]
plt.figure()
plt.plot(df_kappaEuler["kappa"], df_kappaEuler["result"], label= "Euler")
plt.plot(df_kappaAlmost["kappa"], df_kappaAlmost["result"], label= "Almost")
plt.xlabel("kappa")
plt.ylabel("result")
plt.grid(True, which="both")
plt.title("fluctuating $\\kappa$ with fixed $\\theta = 0.11$, $\\sigma = 0.3$ and $\\rho = 0.3$")
plt.legend()
fig = plt.gcf()
fig.savefig('kappa.pgf')
plt.show()
```

print for fluctuating sigma

```python
df_sigma= df[(df["theta"]==0.110) & (df["kappa"]==0.6) & (df["rho"]== 0.3) & (df['dt']<=0.0002)]
df_sigmaEuler= df_sigma[df_sigma["euler"] == 1]
df_sigmaAlmost= df_sigma[df_sigma["almost_exact"] == 1]
plt.figure()
plt.plot(df_sigmaEuler["sigma"], df_sigmaEuler["result"], label= "Euler")
plt.plot(df_sigmaAlmost["sigma"], df_sigmaAlmost["result"], label= "Almost")
plt.xlabel("sigma")
plt.ylabel("result")
plt.grid(True, which="both")
plt.title("fluctuating $\\sigma$ with fixed $\\kappa = 0.6$, $\\theta = 0.11$ and $\\rho = 0.3$")
plt.legend()
fig = plt.gcf()
fig.savefig('sigma.pgf')
plt.show()
```

print for fluctuating theta

```python
df_theta= df[(df["sigma"]==0.3) & (df["kappa"]==0.6) & (df["rho"]== 0.3) & (df['dt']<=0.0002)]
df_thetaEuler= df_theta[df_theta["euler"] == 1]
df_thetaAlmost= df_theta[df_theta["almost_exact"] == 1]
plt.figure()
plt.plot(df_thetaEuler["theta"], df_thetaEuler["result"], label= "Euler")
plt.plot(df_thetaAlmost["theta"], df_thetaAlmost["result"], label= "Almost")
plt.xlabel("theta")
plt.ylabel("result")
plt.grid(True, which="both")
plt.title("fluctuating $\\theta$ with fixed $\\kappa = 0.6$, $\\sigma = 0.3$ and $\\rho = 0.3$")
plt.legend()
fig = plt.gcf()
fig.savefig('theta.pgf')
plt.show()
```

```python
df_dt= df[(df["kappa"] == 0.6) & (df["theta"]==0.110) & (df["sigma"]==0.3) & (df["rho"]== 0.3)]
df_Euler= df_dt[(df_dt["euler"] == 1) & (df_dt["dt"]<0.3)& (df_dt["dt"]>0.005)]
df_Almost= df_dt[(df_dt["almost_exact"] == 1) & (df_dt["dt"]<0.3) & (df_dt["dt"]>0.005)]
plt.figure()
plt.plot(df_Euler["dt"], df_Euler["time"], label= "Euler")
plt.plot(df_Almost["dt"], df_Almost["time"], label= "Almost")
plt.xlabel("dt")
plt.ylabel("execution time (ms)")
plt.grid(True, which="both")
# plt.title("execution time")
plt.legend()
fig = plt.gcf()
fig.savefig('execution.pgf')
plt.show()
```
